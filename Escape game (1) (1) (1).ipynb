```python
#@author: ralph.khoury
import random

""" 
My game is based on Escape plan movie released in 2013 where Sylvester Stallone had the mission to escape one 
of the most guarded high tech prisons in the world.
The mission consists of 4 stages in order to bust out of the prison facility:
Round 1: decide at what time to start the process
Round 2: choose the best way to escape 
Round 3: you need to get out within 20 seconds, counter activated
Round 4: transportation 

"""
#the story begins
#define the input
def start():
    print("""
      After being attacked by a bunch of bad guys, the hero wakes up in a prison cell, he had been kidnapped
      by the ennemies and thrown away in the most guarded prison in the world.
      My game is based on Escape plan movie released in 2013 where Sylvester Stallone had the mission to escape one 
      of the most guarded high tech prisons in the world. 
      """)
#user input    
    input(prompt = "the story seems interesting, now click to continue")
    
    
#define the plan and choose the good timing to move    
def the_plan():
    print("""
    the hero has to choose the right time to move:
    1) midnight
    2) 4pm
    3) 9am
    """)
    choice = input(prompt = ">")

    choice_lower=choice.lower()
        
#the choices we have to escape using if, conditional statement 
    
    if "1" in choice or "midnight" in choice:
        print("great job, guards are sleeping you can resume your plan")
        input(prompt = "press enter to continue")


    elif "2" in choice or "4" in choice or "pm" in choice:
        print("the guards maybe taking a nap, not an easy task")
        input(prompt = "press enter to continue")
    
    
    elif "3" in choice or "9" in choice or "am" in choice:
        print("you have been busted, the guards caught you sneeking out")
        input(prompt = "your game is over now, try again soon")

# adding an else close          
    else:
        print("something is wrong")


    
#defining a variable
def perfect_timing():
    print("""now that you chose the right time to escape you need to 
    figure out the best way to do so.""")
          
    

"""
Now you are about to start the escape process, you need to choose
the best way out

"""


#declaring a list
def escape_idea():
    escape_list = ["digging", "climb_ventilation_system", "unlock_prison_door"]
#choosing a way to escape
    escape=input(prompt = """ escape_way:
                 1.digging
                 2.climb_ventilation_system
                 3.unlock_prison_door """)    
    escape_lower=escape.lower()
#digging could be a solution
    if escape_lower in escape_list [0]:
        print("digging needs a lot of time, not always a good idea")
        time = input(prompt = "how much time do you think will the digging take?\n")
        print(f"""oh great {time} is great, you have enough time
        to dig your way out
        """)
        
#climbing is the easiest way out 

    elif escape_lower in escape_list [1]:
        print("this seems the best and most efficient idea, well done")
        input(prompt = "now press any key to continue")

#unlocking prison cell is not a good idea  

    elif escape_lower in escape_list [2]:
        print("prison alarm has been activated, your escape plan is over")
        input(prompt = 'Press any key to exit\n')
# adding an else close 
    else:
        print("I did not understand please try again")

""" Now that you are out of your cell you need to choose which route 
to take"""
#counter should be less than 20 to leave
def time_to_go():
    
    counter= random.randint(1,20)
#defining integer number    
    integer =int( input(prompt = "how much time do you need to outside the cell? \n"))

#counter to escape should be less than 20    
    while integer <= 20 and counter!=1:
        
        if counter<=20:
        
            print("you were fast enough, great escape")
        
            counter-=1
            break
# adding an else close        
        else:
            
            print("you got caught")
            
            exit()
        
        
#if counter is greater than 20 the game is over        


            
    
""" Congratulations !!! you are free to go, now choose the transporation mode """


#defining transportation, declaring list
def vehicle_list():
    vehicle_list = [["train",0],["boat",1], ["car", 0], ["plane", 0]]
    
#developing for loop
    for j in vehicle_list:
#conditional statement     
        if j[1] == 0:
            print("vehicle chosen is not safe, you failed!!!")
        
        elif j[1] == 1: 
            print("your vehicle is safe to escape")
            
            break
            
    

start()
the_plan()
perfect_timing()
escape_idea()
time_to_go()

vehicle_list()
```
